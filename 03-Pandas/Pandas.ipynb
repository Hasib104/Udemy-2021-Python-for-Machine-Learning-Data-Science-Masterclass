## Merge Method (Inner, Outer, Left or Right Merge)

```python
help(pd.merge)
```

```python
# 1. Inner Merge; Inner join means matches in both tables. 
#on declares the primary key
```

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bob','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bob']})
```

```python
registrations
```

```python
logins
```

```python
#The order of the dataframes doesnt matter in inner join
```

```python
pd.merge(registrations,logins,how='inner',on='name')
```

```python
pd.merge(logins,registrations,how='inner',on='name')
```

```python
# 2. Left and right Merge; 
#left join means showing everything from left + matches in both tables.
#Right join means showing everything from right + matches in both tables.
#order matters in left and right join tables
```

```python
pd.merge(left=registrations,right=logins,how='left',on='name')
```

```python
pd.merge(left=registrations,right=logins,how='right',on='name')
```

```python
# 3. Outer Merge
# Outer join means showing everything from both tables
# order doesnt matter in outer join but sorting will be different
```

```python
pd.merge(registrations,logins,how='outer',on='name')
```

```python
# Joining index 
```

```python
pd.pivot_table(df,index=['Account Manager','Contact','Product'],values=['Sale Price'],aggfunc='sum')
```

```python
#multi-level indexes, columns pivot_table()
```

```python
pd.pivot_table(df,index=['Account Manager','Contact'],values=['Sale Price'],columns=['Product'],aggfunc='sum')
```

```python
pd.pivot_table(df,index=['Account Manager','Contact'],values=['Sale Price'],columns=['Product'],aggfunc='sum',fill_value=0)
```

```python
pd.pivot_table(df,index=['Account Manager','Contact'],values=['Sale Price'],columns=['Product'],aggfunc=[np.sum],fill_value=0)
```

```python
pd.pivot_table(df,index=['Account Manager','Contact'],values=['Sale Price'],columns=['Product'],aggfunc=[np.sum,np.mean],fill_value=0)
```

```python
pd.pivot_table(df,index=['Account Manager','Contact','Product'],values=['Sale Price'],aggfunc=[np.sum],fill_value=0,margins=True)
```
