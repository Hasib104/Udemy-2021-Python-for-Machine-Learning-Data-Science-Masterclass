#### Adding another set of axes to the Figure

```python
fig = plt.figure()
#Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
#Small axes
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])

plt.show()
```

```python
fig = plt.figure()
#Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
#Small axes
axes2 = fig.add_axes([0.2,0.3,0.5,0.5])

plt.show()
```

```python
fig = plt.figure()
#Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
#Small axes
axes2 = fig.add_axes([0.2,0.3,0.1,0.1])

plt.show()
```

```python
fig = plt.figure()
#Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
#Small axes
axes2 = fig.add_axes([0.2,0.3,0.25,0.25])

plt.show()
```

```python
fig = plt.figure()
#Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
#Small axes
axes2 = fig.add_axes([0.2,0.3,0.25,0.25])
axes2.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
#Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)

axes1.set_xlabel('A')
axes1.set_ylabel('B')

axes1.set_title('Power of 4')

#Small axes
axes2 = fig.add_axes([0.2,0.3,0.25,0.25])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)

axes2.set_xlabel('A')
axes2.set_ylabel('B')

axes2.set_title('Zoomed in')

plt.show()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)

axes[0][1].plot(x,y)
axes[0][1].set_xlabel('X label 0,1')

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y label 1,0')

axes[1][1].plot(a,b)
axes[1][1].set_title('Title 1,1')

fig.subplots_adjust(wspace=0.4,hspace=0.8)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)

axes[0][1].plot(x,y)
axes[0][1].set_xlabel('X label 0,1')

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y label 1,0')
axes[1][0].set_xlim(2,6)

axes[1][1].plot(a,b)
axes[1][1].set_title('Title 1,1')

fig.subplots_adjust(wspace=0.4,hspace=0.8)
```

```python
#fig.suptitle means title for all the figures
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)

axes[0][1].plot(x,y)
axes[0][1].set_xlabel('X label 0,1')

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y label 1,0')
axes[1][0].set_xlim(2,6)

axes[1][1].plot(a,b)
axes[1][1].set_title('Title 1,1')

fig.suptitle('Figure Level',fontsize=16)

fig.subplots_adjust(wspace=0.4,hspace=0.8)
```

```python
#Using figsize and dpi for sizing and quality
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(6,6),dpi=100)
axes[0][0].plot(x,y)

axes[0][1].plot(x,y)
axes[0][1].set_xlabel('X label 0,1')

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y label 1,0')
axes[1][0].set_xlim(2,6)

axes[1][1].plot(a,b)
axes[1][1].set_title('Title 1,1')

fig.suptitle('Figure Level',fontsize=16)

fig.subplots_adjust(wspace=0.4,hspace=0.8)

fig.savefig('new_subplots.png',bbox_inches='tight')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label='X vs X')
ax.plot(x,x**2,label='X vs X**2');

ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label='X vs X')
ax.plot(x,x**2,label='X vs X**2');

ax.legend(loc=4)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label='X vs X')
ax.plot(x,x**2,label='X vs X**2');

ax.legend(loc=(0.5,0.5))
```

## Styling : Colors and Styles

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='blue');
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='red');
```

```python
#Adding RGB Hex Code
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e6c017');
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e6c017',label='x vs x')
ax.plot(x,x+1,color='red',label='x vs x+1');
```

```python
#Adding linewidth/lw
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e6c017',lw=5)
ax.plot(x,x+1,color='red',linewidth=0.5);
```

```python
#Adding Line Styles
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e6c017',lw=5,ls='--')
ax.plot(x,x+1,color='red',linewidth=0.5,linestyle='-');
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e6c017',lw=5,ls=':')
ax.plot(x,x+1,color='red',linewidth=0.5,linestyle='-');
```

```python
#Adding custom line designs

```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,x,color='#e6c017',lw=5,ls=':')
lines = ax.plot(x,x+1,color='red',linewidth=5);
```

```python
type(lines)
```

```python
lines
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,x,color='#e6c017',lw=5,ls=':')
lines = ax.plot(x,x+1,color='red',linewidth=5)
lines[0].set_dashes([5,2,10,2]);
#5 dash,2 space,10(5*2)dash,2 space
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,x,color='#e6c017',lw=5,ls=':')
lines = ax.plot(x,x+1,color='red',linewidth=5)
lines[0].set_dashes([1,1,1,2,3,5])
```
